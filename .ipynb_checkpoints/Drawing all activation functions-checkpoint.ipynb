```python
import numpy as np
import matplotlib.pyplot as plt
import activation_functions as af
```

```python
x = np.arange(-1.5, 1.5, 0.01)
y1 = af.linear(x)
y2 = af.binary_threshold(x)
y3 = af.sigmoid(x)
y4 = af.tanh(x)
y5 = af.relu(x)
```

```python
plt.figure(1)
plt.plot(x, y1, 'c',
        x, y2, 'g',
        x, y3, 'b',
        x, y4, 'r',
        x, y5, 'm')

plt.title('Activation Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

# Generalization of the interval

```python
import numpy as np
import matplotlib.pyplot as plt
import activation_functions as af

def af_plot(begin, end, save_fig_name=''):
    %matplotlib inline 
    x = np.arange(begin, end, 0.01)
    y1 = af.linear(x)
    y2 = af.binary_threshold(x)
    y3 = af.sigmoid(x)
    y4 = af.tanh(x)
    y5 = af.relu(x)
    
    plt.figure(1)
    
    plt.plot(x, y1, 'c',
            x, y2, 'g',
            x, y3, 'b',
            x, y4, 'r',
            x, y5, 'm')
    
    plt.title('Activation Functions')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    
    if save_fig_name != '':
        plt.savefig('images/' + save_fig_name)
    
    plt.show()
```

# 1. Interval [-1, 1]

```python
af_plot(-1,1, 'af_1.png')
af_plot(-1,1, 'af_1.pdf')
```

# 2. Interval [-2, 2]

```python
af_plot(-2,2, 'af_2.png')
af_plot(-2,2, 'af_2.pdf')
```

# 3. Interval [-5, 5]

```python
af_plot(-5,5, 'af_5.png')
af_plot(-5,5, 'af_5.pdf')
```

# 4. Interval [-10, 10]

```python
af_plot(-5,5, 'af_10.png')
af_plot(-5,5, 'af_10.pdf')
```
