```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)
```

```python
plt.figure(1)

plt.plot(x, y, 'r')

plt.title('sigmoid')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.savefig('images/' + 'sigmoid_10.png')
plt.savefig('images/' + 'sigmoid_10.pdf')

plt.show()
```
