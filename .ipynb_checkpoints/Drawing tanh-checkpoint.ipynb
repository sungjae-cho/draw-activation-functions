```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.arange(-10.0, 10.0, 0.1)
y = tanh(x)
```

```python
plt.figure(1)

plt.plot(x, y, 'r')

plt.title('tanh')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.savefig('images/' + 'tanh_10.png')
plt.savefig('images/' + 'tanh_10.pdf')

plt.show()
```
