```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-10.0, 10.0, 0.1)
y = relu(x)
```

```python
plt.figure(1)

plt.plot(x, y, 'r')

plt.title('ReLU (Rectified Linear Unit)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.savefig('images/' + 'ReLU_10.png')
plt.savefig('images/' + 'ReLU_10.pdf')

plt.show()
```
