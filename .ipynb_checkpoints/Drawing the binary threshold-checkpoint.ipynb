```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binary_threshold(x):
    x[x > 0] = 1
    x[x <= 0] = 0
    return x
```

```python
x = np.arange(-10.0, 10.0, 0.1)
y = binary_threshold(x)
```

```python
plt.figure(1)

plt.plot(x, y, 'r')

plt.title('binary threshold')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.savefig('images/' + 'binary_threshold_10.png')
plt.savefig('images/' + 'binary_threshold_10.pdf')

plt.show()
```
